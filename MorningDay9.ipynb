```python
#import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset
df=pd.read_csv("crop.csv")
#Display the first few rows of the dataset
print(df.head())
```

```python
#Data cleaning
#check for missing values
print(df.isnull().sum())
```

```python
#check for duplicates
print(df.duplicated().sum())
```

```python
#check for wrong format
print(df.dtypes)
```

```python
#check for wrong data
print(df['ph'].unique())
```

```python
#Exploratory data analysis
#summary statistics
print(df.describe())
```

```python
#check for null/missing values using heatmap    
import seaborn as sbn
sbn.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
#second method
sbn.heatmap(df.isnull(), cmap='coolwarm')
plt.show()
```

```python
#distribution of temperature and ph
print(df['temperature'].value_counts())
```

```python
#method 
sbn.histplot(df['temperature'], bins=30,color='pink', kde=True)
plt.title('Temperature distribution')
plt.show()
```

```python
sbn.histplot(df['ph'], color='green', bins=30, kde=True)
plt.title('ph distribution')
plt.show()
```

```python
#check if the dataset is balanced or not
sbn.countplot(y='label', data=df, palette='plasma_r')
plt.title('Label distribution')
plt.show()
```

```python
#phosphorus level 
sbn.boxplot(y='label', x='ph', data=df, palette='plasma_r')
plt.title('Phosphorus level')
plt.show()
```

```python
#phosphorus level 
sbn.boxplot(y='label', x='P', data=df[df['rainfall']>150])
plt.title('Phosphorus level')
plt.show()
```
